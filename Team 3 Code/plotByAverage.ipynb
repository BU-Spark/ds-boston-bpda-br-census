```python
import numpy as np
import pandas as pd
```

```python
xlsx=pd.ExcelFile("data/Brazilian Immigrants - typo fixed.xlsx")
# state_code=pd.read_excel(xlsx,'State code',index_col='FIPS Code')
state_code=pd.read_excel(xlsx,'State code')
state_code.head()
```

```python
column_name = pd.Series(state_code['State Abbreviation'].values, index=state_code['FIPS Code'])
column_dict=column_name.to_dict()
# print(column_name)
print(column_dict)
```

```python
def read_year(year:str):
    xlsx=pd.ExcelFile("data/Brazilian Immigrants - modified.xlsx")
    df = pd.read_excel(xlsx, year, header=None)
    df.columns = df.iloc[0]
    df.rename(columns=column_dict,inplace=True)
    # for 2014 data, have to do rename columns like this
    c = df.columns.tolist()
    c[0] = 'Type'
    c[2] = 'Class'
    df.columns = c
    df.iloc[0] = df.columns
    return df

read_year('2014').head()
```

```python
df2005=read_year('2005')
df2006=read_year('2006')
df2007=read_year('2007')
df2008=read_year('2008')
df2009=read_year('2009')
df2010=read_year('2010')
df2011=read_year('2011')
df2012=read_year('2012')
df2013=read_year('2013')
df2014=read_year('2014')
df2015=read_year('2015')
df2016=read_year('2016')
df2017=read_year('2017')
df2018=read_year('2018')
df2019=read_year('2019')

data_dict={2005:df2005,2006:df2006,2007:df2007,2008:df2008,2009:df2009,2010:df2010,2011:df2011,2012:df2012,2013:df2013,2014:df2014,2015:df2015,2016:df2016,2017:df2017,2018:df2018,2019:df2019}
```

```python
for i,v in enumerate(data_dict):
    df=data_dict[v]
    df['mean'] = df.iloc[1:, 3:].mean(axis=1)
    df['sum']= df.iloc[1:, 3:].sum(axis=1)
```

```python
def query_by_type(attr,type,class_=""):
    dict={}
    for i,v in enumerate(data_dict):
        df=data_dict[v]
        if len(class_)>1:
            df=df[df['Class']==class_]
        dict[v]=df.loc[df['Type']==attr,type].values[0]
    return dict
```

```python
data=data_dict[2005]
data=data[data['Class']=='Male']
data.head()
```

```python
data=query_by_type('Proportion','mean','Male')
data
```

```python
from pyecharts import options as opts
from pyecharts.charts import Bar, Timeline
from pyecharts.commons.utils import JsCode
from pyecharts.globals import ThemeType
from pyecharts.charts import Grid, Line


def plot(attr,type,class_=""):
  data=query_by_type(attr,type,class_)
  if len(class_)>0:
    title=class_+" "+attr
  else:
    title=attr
  line = (
    Line()
    .add_xaxis(xaxis_data=data.keys())
    .add_yaxis(
        series_name="",
        y_axis=data.values(),
        symbol="emptyCircle",
        is_symbol_show=True,
        label_opts=opts.LabelOpts(is_show=False),
    )
    .set_global_opts(
      title_opts=opts.TitleOpts(
        title=f"{title} in US"
      ),
      tooltip_opts=opts.TooltipOpts(trigger="axis", axis_pointer_type="cross"),
      yaxis_opts=opts.AxisOpts(
        type_="value",
        axistick_opts=opts.AxisTickOpts(is_show=True),
        splitline_opts=opts.SplitLineOpts(is_show=True)
      ),
      xaxis_opts=opts.AxisOpts(type_="category", boundary_gap=False),
      datazoom_opts=opts.DataZoomOpts(
        range_start=0,
        range_end=100
      ),
      legend_opts=opts.LegendOpts(is_show=True, pos_top=40)
    )
  )
  line.render("general/"+title+".html")
  return line
```

```python
plot('Average Age','mean').render_notebook()
```

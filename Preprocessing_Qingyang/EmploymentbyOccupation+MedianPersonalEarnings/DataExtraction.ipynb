```python
def addYear(df,year,num):
    col_name = df.columns.tolist()
    col_name.insert(0,'Year')
    df=df.reindex(columns=col_name)
    years=[]
    for i in range(num):
        years.append(year)
    df['Year']=years
    return df
mpe_2005=addYear(mpe_2005,2005,2)
mpe_2006=addYear(mpe_2006,2006,2)
mpe_2007=addYear(mpe_2007,2007,2)
mpe_2008=addYear(mpe_2008,2008,2)
mpe_2009=addYear(mpe_2009,2009,2)
mpe_2010=addYear(mpe_2010,2010,2)
mpe_2011=addYear(mpe_2011,2011,2)
mpe_2012=addYear(mpe_2012,2012,2)
mpe_2013=addYear(mpe_2013,2013,2)
mpe_2014=addYear(mpe_2014,2014,2)
mpe_2015=addYear(mpe_2015,2015,2)
mpe_2016=addYear(mpe_2016,2016,2)
mpe_2017=addYear(mpe_2017,2017,2)
mpe_2018=addYear(mpe_2018,2018,2)
mpe_2019=addYear(mpe_2019,2019,2)
```

```python
mpe=pd.concat([mpe_2005,mpe_2006,mpe_2007,mpe_2008,mpe_2009,mpe_2010,mpe_2011,
               mpe_2012, mpe_2013,mpe_2014,mpe_2015,mpe_2016,mpe_2017,mpe_2018,mpe_2019])
mpe.to_csv("Median Personal Earnings.csv")
```

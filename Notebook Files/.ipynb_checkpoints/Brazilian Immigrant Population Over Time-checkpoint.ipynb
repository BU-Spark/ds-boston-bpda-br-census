```python
# Graphs age distribution of a single year given the specific year dataframe and an int year (used for title)
def graph_age_range(year_data, year):
    ages = [year_data[25][age] for age in range(1, 18)]
    age_ranges = [year_data['State Code'][ranges] for ranges in range(1,18)]

    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    ax.bar(age_ranges,ages)
    ax.set_ylim(0,18000)
    ax.set_ylabel('Number of Brazilians')
    ax.set_xlabel('Age Ranges')
    ax.set_title('Age Ranges for Brazilian Immigrant Population in Massachusetts for ' + year)
    plt.setp(ax.get_xticklabels(), rotation=30, horizontalalignment='right', fontsize='x-small')
    plt.show()
    
    return


# Graphs gender distribution of a single year given the specific year dataframe and an int year (used for title)
def graph_gender_distribution(year_data, year):
    genders = [year_data[25][gender] for gender in range(18, 20)]
    gender_ranges = [year_data['State Code'][ranges] for ranges in range(18,20)]

    total = year_data[25][0]
    
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    ax.bar(gender_ranges,genders, width = 0.5)
    ax.set_ylim(0,50000)
    for p in ax.patches:
        width = p.get_width()
        height = p.get_height()
        x, y = p.get_xy() 
        ax.annotate('{:.0%}'.format(height/total), (p.get_x()+.5*width, p.get_y() + height + 0.01), ha = 'center')
    
    ax.set_ylabel('Number of Brazilians')
    ax.set_xlabel('Gender')
    ax.set_title('Distribution of Genders for Brazilian Immigrant Population in Massachusetts for ' + year)
    plt.show()
    
    return

# Graphs the distribution of genders overtime 
def graph_gender_over_time(male_count, female_count, years):
    # data to plot
    n_groups = len(male_count)

    # create plot
    fig, ax = plt.subplots()
    index = np.arange(n_groups)
    bar_width = 0.35
    opacity = 0.8

    plt.bar(index, male_count, bar_width,
        alpha=opacity,
        color='b',
        label='Men')

    plt.bar(index + bar_width, female_count, bar_width,
        alpha=opacity,
        color='r',
        label='Women')
    
    plt.xlabel('Years')
    plt.ylabel('Population')
    plt.title('Female and Male Brazilian Population in Massachusetts Over Time')
    plt.xticks(index + bar_width, years)
    plt.legend()

    plt.tight_layout()
    plt.show()


    return 
```

```python
#Graph different types of marital status among the Brazilian population over time
def graph_marital_status(data, years, title):
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    ax.bar(years, data)
    ax.set_ylim(0,max(data)+500)
    ax.set_ylabel('Number of Brazilians')
    ax.set_xlabel('Year')
    ax.set_title(title)
    plt.setp(ax.get_xticklabels(), rotation=30, horizontalalignment='right', fontsize='x-small')
    plt.show()

years = [2005,2006,2007,2008,2009,2010,2011]
graph_marital_status(married_count, years, 'Number of Brazilians Married Between 2005-2011')
graph_marital_status(widowed_count, years, 'Number of Brazilians Widowed Between 2005-2011')
graph_marital_status(divorced_count, years, 'Number of Brazilians Divorced Between 2005-2011')
graph_marital_status(seperated_count, years, 'Number of Brazilians Seperated Between 2005-2011')
graph_marital_status(never_or_under_fifteen_count, years, 'Number of Brazilians Never Married or Under 15 Between 2005-2011')
```

# Employment Type

```python
private_wage = []
government = []
self_employed_notincorportated = []
self_employed_incorportated = []

for year in year_tuple:
    private_wage.append(year[25][38])
    government.append(year[25][39])
    self_employed_notincorportated.append(year[25][40])
    self_employed_incorportated.append(year[25][41])
```

```python
# Graphs the distribution of employment type overtime 
def employment_type_over_time(private_wage, government, self_employed_notincorportated, 
                              self_employed_incorportated, years):
    # data to plot
    n_groups = len(private_wage)

    # create plot
    fig, ax = plt.subplots()
    index = np.arange(n_groups)
    bar_width = 0.2
    opacity = 1

    plt.bar(index, private_wage, bar_width,
        alpha=opacity,
        color='b',
        label='Private Wage and Salary Workers')

    plt.bar(index + bar_width, government, bar_width,
        alpha=opacity,
        color='r',
        label='Government Workers')
    
    plt.bar(index + 2*bar_width, self_employed_notincorportated, bar_width,
        alpha=opacity,
        color='g',
        label='Self employed Not incorporated')
    
    plt.bar(index + 3*bar_width, self_employed_incorportated, bar_width,
        alpha=opacity,
        color='y',
        label='Self employed Incorporated')
    
    plt.xlabel('Years')
    plt.ylabel('Population')
    plt.title('Distribution of Employment Type Among the Brazilian Population in Massachusetts Over Time')
    plt.xticks(index + bar_width, years)
    plt.legend()

    plt.show()

employment_type_over_time(private_wage, government, self_employed_notincorportated, 
                          self_employed_incorportated, years)
```

```python
# Graphs the distribution of employment type not including private wage and salary workers overtime 
def employment_type_over_time_without_private(government, self_employed_notincorportated, 
                              self_employed_incorportated, years):
    # data to plot
    n_groups = len(government)

    # create plot
    fig, ax = plt.subplots()
    index = np.arange(n_groups)
    bar_width = 0.2
    opacity = 1

    plt.bar(index, government, bar_width,
        alpha=opacity,
        color='r',
        label='Government Workers')
    
    plt.bar(index + bar_width, self_employed_notincorportated, bar_width,
        alpha=opacity,
        color='g',
        label='Self employed Not incorporated')
    
    plt.bar(index + 2*bar_width, self_employed_incorportated, bar_width,
        alpha=opacity,
        color='y',
        label='Self employed Incorporated')
    
    plt.xlabel('Years')
    plt.ylabel('Population')
    plt.title('Distribution of Employment Type Among the Brazilian Population in Massachusetts Over Time Not Including Private Wage and Salary Workers')
    plt.xticks(index + bar_width, years)
    plt.legend()

    plt.show()

employment_type_over_time_without_private(government, self_employed_notincorportated, 
                                          self_employed_incorportated, years)
```

# Employment

```python
# graph_employment_industry: Graphs employment distribution by industry of a single year
# Parameters: 
#     df: the dataframe used
#     year: an int year
def graph_employment_industry(df, year):
    
    data = [df[25][data] for data in range(42, 56)]
    features = [df['State Code'][feature] for feature in range(42, 56)]

    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    ax.bar(features,data)
    ax.set_ylim(0, 18000)
    ax.set_ylabel('Number of Brazilians')
    ax.set_xlabel('Industry Types')
    ax.set_title('Industry Types for Brazilian Immigrant Population in Massachusetts for ' + year)
    plt.setp(ax.get_xticklabels(), rotation=30, horizontalalignment='right', fontsize='x-small')
    plt.show()
    
    return

# graph_employment_occupation: Graphs employment distribution by occupation of a single year
# Parameters: 
#     df: the dataframe used
#     year: an int year
def graph_employment_occupation(df, year):
    
    data = [df[25][data] for data in range(56, 66)]
    features = [df['State Code'][feature] for feature in range(56, 66)]

    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    ax.bar(features,data)
    ax.set_ylim(0, 40000)
    ax.set_ylabel('Number of Brazilians')
    ax.set_xlabel('Occupation Types')
    ax.set_title('Occupation Types for Brazilian Immigrant Population in Massachusetts for ' + year)
    plt.setp(ax.get_xticklabels(), rotation=30, horizontalalignment='right', fontsize='x-small')
    plt.show()
    
    return

def line_plot_employment(df, begin_range, end_range, title):
    data = []
    features_legend = []
    
    fig = plt.figure(figsize=(12, 7))
    ax = fig.add_axes([0,0,1,1])
    ax.set_ylim(0,21000)
    ax.set_ylabel('Number of Brazilians')
    ax.set_xlabel('Years')
    ax.set_title(title)
    plt.setp(ax.get_xticklabels(), rotation=30, horizontalalignment='right', fontsize='x-small')
    
    for feature in range (begin_range, end_range):
        data.append(df.loc[feature].tolist())
        features_legend.append(features_list[feature])
    
    for feature in range(len(data)):
        plt.plot(years, data[feature], label = features_legend[feature], color = colors[feature])
    
    plt.legend(loc="upper right", prop={'size': 8})
    plt.show()
```

# Poverty Status

```python
below_poverty_count = []
above_poverty_count = []
not_determined_count = []

for year in year_tuple:
    below_poverty_count.append(year[25][66])
    above_poverty_count.append(year[25][67])
    not_determined_count.append(year[25][68])
```

```python
# Graphs the distribution of Poverty overtime 
def poverty_over_time(below_poverty_count, above_poverty_count, not_determined_count, years):
    # data to plot
    n_groups = len(below_poverty_count)

    # create plot
    fig, ax = plt.subplots()
    index = np.arange(n_groups)
    bar_width = 0.25
    opacity = 0.8

    plt.bar(index, below_poverty_count, bar_width,
        alpha=opacity,
        color='r',
        label='Individuals Below Poverty')

    plt.bar(index + bar_width, above_poverty_count, bar_width,
        alpha=opacity,
        color='g',
        label='Individuals Above Poverty')

    plt.bar(index + 2 * bar_width, not_determined_count, bar_width,
        alpha=opacity,
        color='b',
        label='Not Determined')
    
    plt.xlabel('Years')
    plt.ylabel('Population')
    plt.title('Distribution of Poverty Among the Brazilian Population in Massachusetts Over Time')
    plt.xticks(index + bar_width, years)
    plt.legend()

    plt.show()

poverty_over_time(below_poverty_count, above_poverty_count, not_determined_count, years)
```

# Median Personal Earnings

```python
median_personal_earnings = []
median_personal_earnings_excl_neg_and_zero = []

for year in year_tuple:
    median_personal_earnings.append(year[25][69])
    median_personal_earnings_excl_neg_and_zero.append(year[25][70])
```

```python
# Graphs the distribution of Median Personal Earnings
def median_personal_earnings_over_time(median_personal_earnings, median_personal_earnings_excl_neg_and_zero, years):
    # data to plot
    n_groups = len(median_personal_earnings)

    # create plot
    fig, ax = plt.subplots()
    index = np.arange(n_groups)
    bar_width = 0.3
    opacity = 0.8

    plt.bar(index, median_personal_earnings, bar_width,
        alpha=opacity,
        color='r',
        label='Median Personal Earnings')

    plt.bar(index + bar_width, median_personal_earnings_excl_neg_and_zero, bar_width,
        alpha=opacity,
        color='b',
        label='Median Personal Earnings Excluding 0 and Negative')
    
    plt.xlabel('Years')
    plt.ylabel('Population')
    plt.title('Distribution of Median Personal Earnings Among the Brazilian Population in Massachusetts Over Time')
    plt.xticks(index + bar_width, years)
    plt.legend()

    # manually set axis limits because of legend blocking the data
    plt.axis([-0.5, 6.75, 0, 31000])

    plt.show()

median_personal_earnings_over_time(median_personal_earnings, median_personal_earnings_excl_neg_and_zero, years)
```

# Correlations

The following cells are used to find the correlations between different features. 
1. The first step to finding the correlations is preparing the dataframe to be used. 
    - The dataframe 'ma' contains the data regarding Massachusetts for all features from the year 2005-2011.
    - In order to extract data for a different state, change the '25' to the state code of the desired state. State  code can be found in the project Google Drive provided by clients.
2. The second step is slicing the dataframe created from step one, into different features. 
    - The row number for each feature can be found as comments in the features_list at the beginning of the file.
    - When slicing the pandas dataframe in step one, the end point is inclusive. 
    - When slicing the features_list created at the beginning of the file, the end point is exclusive. 
    - In order to create feature dataframes for another state, use the respective dataframe in step one instead of 'ma'.
3. Choose 2 features to use for the heatmap function. The heatmap function will return the correlation heatmap for the features chosen. 

```python
# Extract data for Massachusetts for each year
ma_2005 = year_2005[25]
ma_2006 = year_2006[25]
ma_2007 = year_2007[25]
ma_2008 = year_2008[25]
ma_2009 = year_2009[25]
ma_2010 = year_2010[25]
ma_2011 = year_2011[25]

# Concatenate all Massachusetts data into one dataframe
ma = pd.concat([ma_2005, ma_2006, ma_2007, ma_2008, ma_2009, ma_2010, ma_2011], axis=1)

# Rename the columns to their respective years
ma = ma.set_axis(['2005','2006', '2007','2008' ,'2009', '2010', '2011'], axis=1, inplace=False)
```

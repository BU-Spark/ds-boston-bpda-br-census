```python
def getUnemploymentGender(data, years, index_male, index_female):
    unemployment_male = {}
    unemployment_female = {}
    for year in years:
      unemployment_male[year] = data[year][index_male]
      unemployment_female[year] = data[year][index_female]
    return unemployment_male, unemployment_female

unemploymentMale, unemploymentFemale = getUnemploymentGender(data, years, 36, 37)
```

```python
def getLaborForceGender(data, years, index_male, index_female):
    labor_male = {}
    labor_female = {}
    for year in years:
      labor_male[year] = data[year][index_male]
      labor_female[year] = data[year][index_female]
    return labor_male, labor_female

labor_male, labor_female = getLaborForceGender(data, years, 34, 35)
```

```python
years = ["2005", "2006", "2007", "2008", "2009", "2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019"]

# equivalent using the read_excel function
data_housing = pd.read_excel(
    "/content/drive/MyDrive/Community Analysis Data/Household TEMP.xlsx", years
)

# index based on state code
# data["year"][index_type]
for item in years:
  data_housing[item] = data_housing[item].drop(labels = data_housing[item].columns[0], axis=1)
  data_housing[item] = data_housing[item].drop(labels = data_housing[item].columns[0], axis=1)
  data_housing[item] = data_housing[item].drop(labels = data_housing[item].columns[0], axis=1)
  data_housing[item] = data_housing[item].drop(labels = data_housing[item].columns[0], axis=1)
  
  data_housing[item] = data_housing[item].transpose()
  data_housing[item].set_index(data_housing[item].columns[0])
  
```

```python
def getPovertyStatusFamilies(data, years, index_above, index_below):
    poverty_below = {}
    poverty_above = {}
    for year in years:
      poverty_above[year] = data[year][index_above]
      poverty_below[year] = data[year][index_below]

    return poverty_below, poverty_above
# above and below poverty threshold
poverty_below, poverty_above = getPovertyStatusFamilies(data_housing, years, 2, 1)
```

```python
def getCrowdedStatus(data, years, index_yes, index_no):
    crowded = {}
    not_crowded = {}
    for year in years:
      crowded[year] = data[year][index_yes]
      not_crowded[year] = data[year][index_no]

    return crowded, not_crowded

crowded, not_crowded = getCrowdedStatus(data_housing, years, 16, 17)
```

```python
def getRenterBurden(data, years, index):
    renter_burden = {}
    for year in years:
      renter_burden[year] = data[year][index]
    return renter_burden

HH_Income_under_15 = getRenterBurden(data_housing, years, 10)
HH_Income_15_20 = getRenterBurden(data_housing, years, 11)
HH_Income_20_25 = getRenterBurden(data_housing, years, 12)
HH_Income_25_30 = getRenterBurden(data_housing, years, 13)
HH_Income_30_35 = getRenterBurden(data_housing, years, 14)
HH_Income_above_35 = getRenterBurden(data_housing, years, 15)
```

```python
# Plot Renter Burden Brackets over time for Massachusetts 
HH_Income_under_15["2005"][25]

def makeBracketYearsList(data, years, state_code):
    bracket_list = []
    for year in years:
        bracket_list += [data[year][state_code]]
    return bracket_list

List_MA_HH_Income_under_15 = makeBracketYearsList(HH_Income_under_15, years, 25)
List_MA_HH_Income_15_20 = makeBracketYearsList(HH_Income_15_20, years, 25)
List_MA_HH_Income_20_25 = makeBracketYearsList(HH_Income_20_25, years, 25)
List_MA_HH_Income_25_30 = makeBracketYearsList(HH_Income_25_30, years, 25)
List_MA_HH_Income_30_35 = makeBracketYearsList(HH_Income_30_35, years, 25)
List_MA_HH_Income_above_35 = makeBracketYearsList(HH_Income_above_35, years, 25)
```

```python
plt.rcParams["figure.figsize"] = (15, 5)
plt.plot(years, List_MA_HH_Income_above_35, '.-',label = "Above 35%")
plt.plot(years, List_MA_HH_Income_30_35, '.-',label = "30%-35%")
plt.plot(years, List_MA_HH_Income_25_30, '.-',label = "25%-30%")
plt.plot(years, List_MA_HH_Income_20_25, '.-',label = "20%-25%")
plt.plot(years, List_MA_HH_Income_15_20, '.-',label = "15%-20%")
plt.plot(years, List_MA_HH_Income_under_15, '.-',label = "Under 15%")


plt.title("Change in Household Income Spent on Rent")
plt.xlabel("Year")
plt.ylabel("Number of people")
plt.legend()
plt.show()
```

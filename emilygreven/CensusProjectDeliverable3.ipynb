```python
def getPovertyStatusFamilies(data, years, index_above, index_below):
    poverty_below = {}
    poverty_above = {}
    for year in years:
      poverty_above[year] = data[year][index_above]
      poverty_below[year] = data[year][index_below]

    return poverty_below, poverty_above
# above and below poverty threshold
poverty_below, poverty_above = getPovertyStatusFamilies(data_housing, years, 2, 1)
```

```python
def getCrowdedStatus(data, years, index_yes, index_no):
    crowded = {}
    not_crowded = {}
    for year in years:
      crowded[year] = data[year][index_yes]
      not_crowded[year] = data[year][index_no]

    return crowded, not_crowded

crowded, not_crowded = getCrowdedStatus(data_housing, years, 16, 17)
```

```python
def getRenterBurden(data, years, index):
    renter_burden = {}
    for year in years:
      renter_burden[year] = data[year][index]
    return renter_burden

HH_Income_under_15 = getRenterBurden(data_housing, years, 10)
HH_Income_15_20 = getRenterBurden(data_housing, years, 11)
HH_Income_20_25 = getRenterBurden(data_housing, years, 12)
HH_Income_25_30 = getRenterBurden(data_housing, years, 13)
HH_Income_30_35 = getRenterBurden(data_housing, years, 14)
HH_Income_above_35 = getRenterBurden(data_housing, years, 15)
```

```python
def getFeatureData(data, years, index):
    feature_dict = {}
    for year in years:
      feature_dict[year] = data[year][index]
    return feature_dict

def makeFeatureDataListState(data, years, feature_index, state_code):
    feature_data = getFeatureData(data, years, feature_index)
    data_list = []
    for year in years:
      data_list += [feature_data[year][state_code]]
    return data_list


MA_df = pd.DataFrame([makeFeatureDataListState(data, years, x, 25) for x in range(71)])
MA_df.columns = years
MA_df = MA_df.transpose()
MA_df.columns = Categories

# TOOK OUT: "Median Household Income", "Median Family Income",
housing_df = pd.DataFrame([makeFeatureDataListState(data_housing, years, x, 25) for x in range(19)])
housing_df.columns = years
housing_df = housing_df.transpose()
housing_df.columns = Categories_housing
```

```python
# Make linear regression Function
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression, LogisticRegression

def makeLinReg(X, y):
  
 
  X_intercept = sm.add_constant(X)
  
  X_train, X_test, y_train, y_test = model_selection.train_test_split(
    X_intercept, y, test_size = 0.5, random_state = 0)

  model = sm.OLS(y_train, X_train)
  results = model.fit()
  print(results.summary())

  X_train, X_test, y_train, y_test = model_selection.train_test_split(
    X, y, test_size = 0.5, random_state = 0)

  # Create linear regression object
  regr = linear_model.LinearRegression()

  # Train the model using the training sets
  regr.fit(X_train, y_train)

  # Make predictions using the testing set
  y_pred = regr.predict(X_test)
  # The coefficients
  print("Coefficients: \n", regr.coef_)
  # The mean squared error
  print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
  # The coefficient of determination: 1 is perfect prediction
  print("Coefficient of determination: %.2f" % r2_score(y_test, y_pred))


  return X_test, y_test, y_pred


X_test, y_test, y_pred = makeLinReg(marital_status[["Married", "Widowed", "Divorced",
]], marital_status["Female"])

X = np.array(marital_status[["Married"]]).reshape(-1, 1)
y = marital_status["Female"]

model = LinearRegression().fit(X, y)

xfit = np.linspace(20000, 70000, len(X))
yfit = model.predict(X)

yline = [model.intercept_+x*model.coef_ for x in xfit]

plt.scatter(X, y)
plt.plot(xfit, yfit)
plt.plot(xfit, yline)
plt.title('Marriage and Females -- Correlation')
plt.xlabel('Married')
plt.ylabel('Female')
plt.show()

```
